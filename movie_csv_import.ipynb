```python
# import data from the csv
import pandas as pd
import numpy as py

csv_file = "netflix_titles.csv"
netflix_file_df = pd.read_csv(csv_file)
netflix_file_df.head()
```

```python
# drop na on file
netflix_file_df_clean = netflix_file_df.dropna()
netflix_file_df_clean.info()
```

```python
# only movies from US
netflix_file_df_clean = netflix_file_df_clean.loc[netflix_file_df_clean["country"] == "United States"]
netflix_file_df_clean.info()
```

```python
# only movies
netflix_file_df_clean = netflix_file_df_clean.loc[netflix_file_df_clean["type"] == "Movie"]
netflix_file_df_clean.info()
```

```python
# movies after 2010
netflix_file_df_clean = netflix_file_df_clean.loc[netflix_file_df_clean["release_year"] >= 2010]
netflix_file_df_clean.info()
```

```python
netflix_file_df_clean.head()
```

```python
# save to a new csv
netflix_file_df_clean.to_csv("netflix_data_clean.csv", header=True)
```
